```python
from pandas import DataFrame, read_csv
from numpy import random, arange
```

```python
base = {
    'ids': arange(1, 1001),
    'idades': random.randint(18, 80, size=1000),
    'sexos': random.choice(['M', 'F'], size=1000),
    'estados_civis': random.choice(['Solteiro(a)', 'Casado(a)', 'Solteiro(a)', 'Divorciado(a)', 'Solteiro(a)'],1000),
    'sinistros': random.randint(0, 5, 1000),
    'premios': random.gamma(2, 1000, size=1000)
}
```

```python
base = DataFrame(base)
base.head()
```

```python
clientes_maior_30anos = base[base['idades'] > 30]
clientes_maior_30anos
```

```python
clientes_sem_sinistro = base.query('sinistros == 0')
clientes_sem_sinistro
```

```python
dados = read_csv('dados.csv', index_col='Unnamed: 0')
dados
```

```python
def classificar_idade(idade):
    if idade >= 18 and idade <= 30:
        return 'Jovem'
    elif idade >= 31 and idade <= 45:
        return 'Adulto'
    else:
        return 'Sênior'
```

```python
dados['Categoria'] = dados['idades'].apply(classificar_idade)
dados
```

```python
def calcular_desconto(pessoa):
    if pessoa['sinistros'] == 0:
        if pessoa['Categoria'] == 'Jovem':
            return pessoa['premios'] * 0.05
        elif pessoa['Categoria'] in ['Adulto', 'Sênior']:
            return pessoa['premios'] * 0.1
    return 0
```

```python
dados['Desconto'] = dados.apply(calcular_desconto, axis=1)
dados
```
